#**Análises Comparativas das Planilhas das Eleições**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#**Aptos à Votar**

```python
Total_Aptos = pd.DataFrame ({
    'Eleitores Aptos': ['125.913.479','135.804.433','142.820.753','147.306.275', '156.454.011'],
    'Ano': [2006, 2010, 2014, 2018, 2022],

})
Total_Aptos['Ano'] = Total_Aptos['Ano'].astype(int)
Total_Aptos['Eleitores Aptos'] = Total_Aptos['Eleitores Aptos'].str.replace('.', '', regex=False).astype(int)
```

**Tabela**

```python
Total_Aptos
```

```python
Média_TotalAptos = int(Total_Aptos['Eleitores Aptos'].mean())
print(f"A média de eleitores aptos entre os anos ordenados é: {Média_TotalAptos} eleitores")
```

**Gráfico**

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Ano', y='Eleitores Aptos', data=Total_Aptos, marker='o')
plt.title('Total de Eleitores Aptos por Ano')
plt.xlabel('Ano')
plt.ylabel('Total de Eleitores Aptos')
plt.grid(True)
plt.show()
```

#**Votos Quantitativos Pelo 1° Turno**

```python
Total_Votos_turno1 = pd.DataFrame({
    'Ano':[2006, 2010, 2014, 2018, 2022],
    'Candidatos':['98.561.346', '101.590.153', '104.023.802', '107.051.495', '118.229.719' ],
    'Votos Brancos Nulos' :['8.797.957','9.603.594', '11.099.081', '10.313.159	', '5.452.653	']
})
Total_Votos_turno1['Candidatos'] = Total_Votos_turno1['Candidatos'].str.replace('.', '', regex=False).astype(int)
Total_Votos_turno1['Votos Brancos Nulos'] = Total_Votos_turno1['Votos Brancos Nulos'].str.replace('.', '', regex=False).astype(int)
Total_Votos_turno1['Votos Quantitativos'] = Total_Votos_turno1['Candidatos'] + Total_Votos_turno1['Votos Brancos Nulos']
```

```python
Total_Votos_turno1
```

```python
Média_BrancoeNulo = int(Total_Votos_turno1['Votos Brancos Nulos'].mean())
print(f"Média votos brancos e nulos entre os anos ordenados é: {Média_BrancoeNulo} votos")
```
